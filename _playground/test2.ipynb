```python
from pathlib import Path

from fractal_uzh_converters.common.image_in_plate_compute_task import (
    image_in_plate_compute_task,
 )
from fractal_uzh_converters.cq3k.convert_cq3k_init_task import (
    CQ3KAcquisitionModel,
    OverwriteMode,
    convert_cq3k_init_task,
 )
```

```python
input_root = Path(
    "/links/shared/scuanalysis/LAF/Sant/P2001_Hierlemann_00013515_with_CV_format/20251027T134435_10X"
 )
output_zarr = Path("/tmp/cq3000_simple_test_ome_zarr")

p_list = convert_cq3k_init_task(
    zarr_dir=str(output_zarr),
    acquisitions=[
        CQ3KAcquisitionModel(
            path=str(input_root),
            acquisition_id=0,
        ),
    ],
    overwrite=OverwriteMode.OVERWRITE,
 )

print(f"Prepared {len(p_list['parallelization_list'])} conversion tasks")
```

```python
assert len(p_list["parallelization_list"]) > 0, "No tasks generated from input path"

for task in p_list["parallelization_list"]:
    results = image_in_plate_compute_task(**task)
    assert "image_list_updates" in results
    assert len(results["image_list_updates"]) == 1

print(f"Conversion completed successfully: {output_zarr}")
```
